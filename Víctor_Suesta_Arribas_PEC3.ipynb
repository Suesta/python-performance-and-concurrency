## Ejercicio 1

En esta serie de ejercicios trabajaremos sobre la eficiencia temporal de los algoritmos.
Analizaremos, optimizaremos y compararemos diferentes implementaciones de un problema sencillo: detectar si una lista contiene elementos duplicados.

El objetivo es entender cómo la elección de un enfoque u otro en la implementación puede modificar de forma notable el rendimiento de un algoritmo.
Recordad que el rendimiento de un algoritmo no solo depende del tamaño de la entrada, sino también de sus valores específicos.
Por este motivo, al analizar la eficiencia se suele considerar el caso peor, es decir, aquel en el que el algoritmo requiere el mayor número de operaciones para completarse.


<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **Ejercicio 1.1 (0.5 puntos)**

En este ejercicio debéis analizar la complejidad temporal del siguiente código y expresar su orden de complejidad.


A partir del siguiente algoritmo, determinad su complejidad en notación O, justificando brevemente vuestra respuesta.

`````python
def has_duplicates(num_list: list[int]) -> bool:
    """
    Checks whether a list of ints contains duplicate values.

    Args:
        num_list (list[int]): List of integers to analyze.

    Returns:
        bool: True if duplicates are found, False otherwise.
    """
    for k in range(len(num_list)):
        for v in range(k + 1, len(num_list)):
            if num_list[k] == num_list[v]:
                return True
    return False
`````

Sea `n = len(num_list)` el tamaño de la lista de entrada.

1. En primer lugar, observo que el **bucle externo**:

   ```python
   for k in range(len(num_list)):
   ```

   recorre todos los índices desde `k = 0` hasta `k = n-1`. Por tanto, se ejecuta `n` veces.

2. Para cada valor de `k`, el **bucle interno**:

   ```python
   for v in range(k + 1, len(num_list)):
   ```

   recorre los índices `v` desde `k+1` hasta `n-1`.

   * Cuando `k = 0`, el bucle interno hace `n-1` iteraciones.
   * Cuando `k = 1`, hace `n-2` iteraciones.
   * …
   * Cuando `k = n-2`, hace 1 iteración.
   * Cuando `k = n-1`, hace 0 iteraciones.

   En el **caso peor** (cuando supongo que no hay duplicados y nunca se ejecuta el `return True`), el número total de iteraciones del bucle interno es:

   `(n-1) + (n-2) + ... + 2 + 1 = n(n-1)/2`

   En cada una de esas iteraciones se realiza, como operación relevante, la comparación:

   ```python
   if num_list[k] == num_list[v]:
   ```

3. Por tanto, en el caso peor el número de comparaciones crece como `n(n-1)/2`, que es una función cuadrática en `n`.
   Si ignoro constantes y términos de menor grado, me quedo con el término dominante `n^2/2`, y la complejidad temporal en notación O grande es:

   `T(n) ∈ O(n^2)`

4. En el **mejor caso**, si los dos primeros elementos ya son iguales, el algoritmo detecta el duplicado en la primera comparación y retorna inmediatamente `True`. En ese escenario, el tiempo de ejecución es básicamente constante (no depende de `n` de forma apreciable), es decir, `T_mejor(n) ∈ O(1)`. Sin embargo, cuando se habla de complejidad de este tipo de algoritmos normalmente se toma como referencia el **caso peor**, que aquí es claramente cuadrático.

En resumen, concluyo que el algoritmo `has_duplicates` tiene **complejidad temporal cuadrática** en el caso peor debido a la presencia de **dos bucles anidados** cuyo número total de iteraciones crece proporcionalmente a `n^2`. Esto encaja con lo que se explica en la Unidad 4, donde la complejidad de un algoritmo se puede deducir observando su estructura de bucles y quedándose con el término de mayor grado al contar pasos.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **Ejercicio 1.2 (0.75 puntos)**

En este ejercicio debéis mejorar el algoritmo `has_duplicates` de manera que reduzca su complejidad temporal. Para ello, cread una nueva función llamada `has_duplicates_improved` e implementad el código correspondiente.

Comprobad que ambas versiones producen el mismo resultado al probarlas con diferentes entradas. Para ello, añadid una verificación adicional usando `assert`. Una vez mejorada, indicad la complejidad temporal del nuevo algoritmo y explicad brevemente en qué consiste la mejora respecto a la versión original.


`````python
def has_duplicates_improved(num_list: list[int]) -> bool:
    """
    Checks whether a list of ints contains duplicate values using
    an auxiliary set to track seen elements.

    Args:
        num_list (list[int]): List of integers to analyze.

    Returns:
        bool: True if duplicates are found, False otherwise.
    """
    seen = set()
    for value in num_list:
        # si el valor ya estaba en el conjunto es que hay duplicado
        if value in seen:
            return True
        # si no estaba lo añadimos para futuras comprobaciones
        seen.add(value)
    # si llegamos hasta aqui es que no hemos encontrado ningun duplicado
    return False


# listas sin duplicados
lista_a = [1, 2, 3, 4, 5]
lista_b = list(range(100))

# listas con duplicados en distintas posiciones
lista_c = [1, 1]
lista_d = [1, 2, 3, 2, 5]
lista_e = [0] * 10

# compruebo que las dos funciones devuelven lo mismo en varios casos
assert has_duplicates(lista_a) == has_duplicates_improved(lista_a)
assert has_duplicates(lista_b) == has_duplicates_improved(lista_b)
assert has_duplicates(lista_c) == has_duplicates_improved(lista_c)
assert has_duplicates(lista_d) == has_duplicates_improved(lista_d)
assert has_duplicates(lista_e) == has_duplicates_improved(lista_e)

# si llega hasta aqui sin saltar ninguna aserción, las versiones son consistentes
print("Comprobación pasada: ambas funciones devuelven los mismos resultados.")
`````

### Versión mejorada de `has_duplicates`

En este ejercicio he implementado una nueva función `has_duplicates_improved` para reducir la complejidad temporal del algoritmo original. La idea que sigo es recorrer la lista una sola vez e ir guardando en un conjunto (`set`) los elementos que ya han aparecido:

* Si el elemento actual **ya está** en el conjunto, entiendo que es un duplicado y devuelvo `True`.
* Si no está, lo añado al conjunto y continúo con el siguiente elemento.
* Si llego al final del recorrido sin encontrar ningún valor repetido, devuelvo `False`.

El código que utilizo es el siguiente (lo tengo en la celda de código anterior):

```python
def has_duplicates_improved(num_list: list[int]) -> bool:
    seen = set()
    for value in num_list:
        if value in seen:
            return True
        seen.add(value)
    return False
```

Para comprobar que ambas versiones se comportan igual, he creado varias listas de prueba (con y sin duplicados, y con los duplicados en distintas posiciones) y he añadido varias sentencias `assert` del tipo:

```python
assert has_duplicates(lista_x) == has_duplicates_improved(lista_x)
```

Si alguna de las dos funciones devolviera un resultado distinto, Python lanzaría una excepción y vería el error al ejecutar la celda. Como no se produce ningún error y al final se imprime el mensaje de confirmación, doy por bueno que las dos implementaciones son consistentes para los casos que he probado.



#### Complejidad temporal y mejora respecto a la versión original

En la **versión original** de `has_duplicates`, hay dos bucles anidados que comparan todos los pares posibles de elementos en el caso peor. Ya he visto en el Ejercicio 1.1 que su complejidad temporal en notación O grande es:

* `T_peor(n) ∈ O(n^2)`

En cambio, en la nueva versión:

* Recorro la lista **una sola vez** con un bucle `for`, así que hago `n` iteraciones.
* En cada iteración realizo:

  * Una comprobación de pertenencia `value in seen`.
  * Una inserción `seen.add(value)` (solo cuando el elemento no estaba).

En Python, las operaciones de pertenencia e inserción sobre un `set` se implementan de forma que su coste medio es constante, es decir, `O(1)` en promedio. Por tanto, el coste total del algoritmo en el caso peor pasa a ser:

* `T_peor(n) ∈ O(n)`

Frente a la versión original, he pasado de un algoritmo **cuadrático** a un algoritmo **lineal**, lo que supone una mejora muy importante cuando el tamaño de la lista `n` es grande.

Esta idea coincide con el ejemplo de la Unidad 4 donde se compara el tiempo de búsqueda de un elemento en un conjunto (`set`) y en una lista (`list`): allí se comenta que el test de pertenencia en un conjunto tiene complejidad constante `O(1)`, mientras que en una lista es lineal `O(n)`, y se ve experimentalmente que el conjunto escala mucho mejor cuando crece el tamaño de los datos.

En este apartado comparo el **tiempo de ejecución** de los dos algoritmos que detectan si una lista
contiene elementos duplicados:

* `has_duplicates_improved` (versión mejorada del ejercicio 1.2, basada en un conjunto).

Para ello, lo que hago es:

La idea que quiero comprobar empíricamente es que:

* La versión original crece de forma **cuadrática** (O(n^2)).
* La versión mejorada crece de forma aproximadamente **lineal** (O(n)), al aprovechar el test de
  pertenencia sobre un `set`, cuyo coste promedio es constante.

Esta forma de trabajo sigue el enfoque de la Unidad 4, donde se propone comparar el tiempo de ejecución
de distintas implementaciones midiendo el tiempo real de ejecución para diferentes tamaños de entrada,
como en el ejemplo de test de pertenencia en conjuntos frente a listas.

`````python
import time
import random
import matplotlib.pyplot as plt

# tamaños de lista
sizes = [200, 600, 1000, 1400, 1800, 5000]


def generar_lista_sin_duplicados(n: int) -> list[int]:
    """
    Genera una lista de 'n' enteros sin valores repetidos.

    La lista se obtiene seleccionando 'n' números distintos de un rango
    suficientemente grande y barajándolos para que el orden no sea tan
    "artificial".

    Args:
        n (int): Tamaño de la lista deseada.

    Returns:
        list[int]: Lista de enteros sin duplicados.
    """
    # uso un rango un poco mas grande por si acaso
    universo = list(range(10 * max(sizes)))
    muestra = random.sample(universo, n)
    random.shuffle(muestra)
    return muestra


def medir_tiempo_medio(func, datos: list[int], repeticiones: int = 5) -> float:
    """
    Mide el tiempo medio de ejecución de una función sobre una lista dada.

    Args:
        func (callable): Función a evaluar (por ejemplo has_duplicates).
        datos (list[int]): Lista de enteros sobre la que se aplica la función.
        repeticiones (int): Número de veces que se repite la llamada para
            suavizar el ruido de las medidas.

    Returns:
        float: Tiempo medio de ejecución en segundos.
    """
    inicio = time.perf_counter()
    for _ in range(repeticiones):
        func(datos)
    fin = time.perf_counter()
    # divido por las repeticiones para obtener el tiempo medio
    return (fin - inicio) / repeticiones


# listas para guardar los tiempos de cada algoritmo
tiempos_original = []
tiempos_mejorado = []

# para reproducibilidad relativa
random.seed(42)

for n in sizes:
    # genero una lista sin duplicados
    lista = generar_lista_sin_duplicados(n)

    # mido tiempo de la versión original
    t_orig = medir_tiempo_medio(has_duplicates, lista, repeticiones=5)
    tiempos_original.append(t_orig)

    # mido tiempo de la versión mejorada
    t_imp = medir_tiempo_medio(has_duplicates_improved, lista, repeticiones=5)
    tiempos_mejorado.append(t_imp)

    # imprimo un pequeño resumen por consola para ver como va
    print(f"n = {n:4d} | original = {t_orig:.6f}s | improved = {t_imp:.6f}s")

# representación gráfica de los resultados
plt.figure(figsize=(7, 4))
plt.plot(sizes, tiempos_original, marker="o", label="has_duplicates (O(n²))")
plt.plot(sizes, tiempos_mejorado, marker="o", label="has_duplicates_improved (O(n))")
plt.xlabel("Tamaño de la lista (n)")
plt.ylabel("Tiempo medio de ejecución (s)")
plt.title("Comparación del tiempo de ejecución según el tamaño de la lista")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
`````



#### Interpretación de los resultados

En la tabla impresa por pantalla y en la gráfica puedo ver que el tiempo de ejecución de
`has_duplicates` (versión original) crece muy rápido con el tamaño de la lista. Para `n = 200` el
tiempo está alrededor de milésimas de segundo, pero para `n = 5000` ya supera las **0.6 s**.
La forma de la curva es claramente **curvada hacia arriba**, típica de un crecimiento cuadrático
(O(n^2)), que es lo que había obtenido al analizar los dos bucles anidados en el ejercicio 1.1.

En cambio, la versión mejorada `has_duplicates_improved` mantiene tiempos casi constantes
en todos los tamaños de lista: los valores están del orden de (10^{-4})–(10^{-3}) segundos, y la
curva aparece prácticamente **aplanada** en la gráfica. Esto es coherente con la complejidad
aproximadamente lineal (O(n)) del algoritmo mejorado, que se basa en ir añadiendo los elementos
a un conjunto y comprobar pertenencia con coste medio constante, de forma análoga al ejemplo de
la Unidad 4 donde se compara la búsqueda en `set` frente a `list`.

Empíricamente, por tanto, para mí se confirma la teoría: para tamaños pequeños las diferencias entre ambos
algoritmos todavía son moderadas, pero a medida que aumenta (n) la versión cuadrática se vuelve
rápidamente mucho más lenta que la versión casi lineal. Este tipo de comparativa es justo el enfoque
que se propone en la Unidad 4: analizar primero la complejidad asintótica y luego comprobarla
mediante medidas de tiempo con diferentes tamaños de entrada, como en el ejemplo de comparación entre `set` y `list`.

En este ejercicio trabajo con el fichero `music_dataset.csv`, que contiene información de usuarios de una plataforma de música en línea. Lo que me piden es calcular el **tiempo medio de reproducción** (`listening_time`) para cada tipo de suscripción (`subscription_type`) implementando **dos versiones** distintas:

Las dos funciones que he implementado cumplen estas condiciones:

* Devuelven **únicamente** un diccionario de la forma
  `{subscription_type (str): media_listening_time (float)}`.
* No imprimen nada por pantalla ni hacen transformaciones extra de los datos, para ajustarme a lo que pide el enunciado.

En concreto, he definido:

* `avg_listening_time_manual(path: str) -> Dict[str, float]`
  Aquí leo el archivo con `open()`, tomo la primera línea como cabecera para localizar las columnas `subscription_type` y `listening_time`, y después recorro el fichero **línea a línea**. Para cada fila válida:

  * convierto el tiempo de escucha a `float`,
  * acumulo sumas y contadores por tipo de suscripción,
  * y al final calculo la media `suma / conteo` para cada tipo.

* `avg_listening_time_pandas(path: str) -> Dict[str, float]`
  En esta versión cargo el CSV completo en un `DataFrame` con `pd.read_csv(path)` y agrupo por `subscription_type` usando `groupby`. Luego calculo la media de `listening_time` con `.mean()` y paso el resultado a diccionario con `.to_dict()` para devolver exactamente el mismo tipo de salida que en la versión manual.

Esta forma de resolver el ejercicio sigue bastante la idea que se comenta en la Unidad 4: primero planteo una solución “a bajo nivel” con bucles y estructuras básicas de Python, y luego aplico la misma lógica usando una librería especializada (`pandas`), que es más declarativa y cómoda cuando se trabaja con datos tabulares y se quieren hacer agregaciones.

* Documentación oficial de `pandas` (secciones de `read_csv`, `DataFrame.groupby` y `Series.mean`).


`````python
def avg_listening_time_manual(path: str) -> Dict[str, float]:
    """
    Implementación A (lectura manual).

    Calcula el tiempo medio de reproducción (listening_time) por tipo de
    suscripción (subscription_type) leyendo el CSV línea a línea con
    funciones básicas de Python.

    Args:
        path (str): Ruta al fichero CSV (music_dataset.csv).

    Returns:
        Dict[str, float]: Diccionario {subscription_type: media_listening_time}.
    """
    # diccionario para acumular las sumas de listening_time
    sumas: Dict[str, float] = {}
    # diccionario para contar cuántos registros hay de cada tipo
    conteos: Dict[str, int] = {}

    with open(path, "r", encoding="utf-8") as f:
        # leo la primera línea como cabecera
        header = f.readline().strip()
        columnas = header.split(",")

        # localizo las columnas que me interesan por nombre
        try:
            idx_sub = columnas.index("subscription_type")
            idx_time = columnas.index("listening_time")
        except ValueError as exc:
            raise ValueError(
                "No se han encontrado las columnas 'subscription_type' "
                "y/o 'listening_time' en la cabecera del CSV."
            ) from exc

        # recorro el resto de líneas una a una
        for linea in f:
            linea = linea.strip()
            if not linea:
                # si hay una línea vacía la salto
                continue

            partes = linea.split(",")

            # pequeño control por si la línea no tiene suficientes columnas
            if len(partes) <= max(idx_sub, idx_time):
                continue

            sub_type = partes[idx_sub]
            time_str = partes[idx_time]

            # intento convertir el tiempo a float; si falla, salto la fila
            try:
                listening_time = float(time_str)
            except ValueError:
                continue

            # inicializo acumuladores si hace falta
            if sub_type not in sumas:
                sumas[sub_type] = 0.0
                conteos[sub_type] = 0

            # acumulo suma y contador
            sumas[sub_type] += listening_time
            conteos[sub_type] += 1

    # calculo la media para cada tipo de suscripción
    resultado: Dict[str, float] = {}
    for sub_type, suma in sumas.items():
        n = conteos[sub_type]
        if n > 0:
            resultado[sub_type] = suma / n

    return resultado


def avg_listening_time_pandas(path: str) -> Dict[str, float]:
    """
    Implementación B (con pandas).

    Calcula el tiempo medio de reproducción (listening_time) por tipo de
    suscripción (subscription_type) utilizando pandas.read_csv y groupby+mean.

    Args:
        path (str): Ruta al fichero CSV (music_dataset.csv).

    Returns:
        Dict[str, float]: Diccionario {subscription_type: media_listening_time}.
    """
    # leo el CSV completo en un DataFrame
    df = pd.read_csv(path)

    # agrupo por tipo de suscripción y calculo la media de listening_time
    medias = (
        df.groupby("subscription_type")["listening_time"]
        .mean()
        .to_dict()
    )

    return medias
`````

`````python
def factorial_sum(n: int):
    """Returns the sum of the factorials of the first n natural numbers.

    Args:
        n (int): Positive integer indicating the upper limit of the factorial sum.

    Returns:
        (int): Sum of factorials from 1! to n!.
    """
    total = 0
    for i in range(1, n + 1):
        fact = 1
        for j in range(1, i + 1):
            fact *= j
        total += fact
    return total

# factorial_sum(3500)
`````

La función calcula la suma de los factoriales desde (1!) hasta (n!) con dos bucles anidados:

* El bucle externo recorre `i` de 1 a `n`.
* Para cada `i`, el bucle interno recalcula el factorial de `i` desde cero:

En este ejercicio mejoro el código de `factorial_sum` con el objetivo de reducir su tiempo de ejecución. A partir del análisis del Ejercicio 3.1, he visto que el cuello de botella está en el bucle interno:

```python
fact = 1
for j in range(1, i + 1):
    fact *= j
```

Para cada valor de `i` vuelvo a calcular desde cero el factorial de `i`, aunque en realidad podría obtenerlo a partir del factorial anterior. La idea de la versión mejorada es precisamente reutilizar el factorial ya calculado para el valor anterior:

* Mantengo una variable `fact` que guarda el factorial actual.
* En cada iteración del bucle externo actualizo `fact` como `fact *= i`, con lo que paso de `i-1` a `i`.
* Luego acumulo ese factorial en `total`.

De este modo:

* En la versión original se realizan del orden de (1 + 2 + \dots + n = n(n+1)/2) multiplicaciones, por lo que la complejidad temporal es **O(n²)**.
* En la nueva versión solo hago **una multiplicación por iteración** del bucle externo, así que el número total de multiplicaciones es proporcional a (n) y la complejidad pasa a ser **O(n)**.

Además, compruebo que ambas funciones (`factorial_sum` y `factorial_sum_optimized`) devuelven el mismo resultado para varios valores de prueba, y comparo sus tiempos de ejecución con `%time` para ver la mejora en la práctica.

En resumen, la mejora consiste en **evitar recalcular desde cero cada factorial** y reutilizar el factorial anterior, lo que me permite reducir la complejidad de cuadrática a lineal.


`````python
# Versión optimizada de factorial_sum y comparación de tiempos

def factorial_sum_optimized(n: int) -> int:
    """
    Returns the sum of the factorials of the first n natural numbers
    using an optimized O(n) algorithm.

    Args:
        n (int): Positive integer indicating the upper limit of the factorial sum.

    Returns:
        int: Sum of factorials from 1! to n!.
    """
    if n < 0:
        raise ValueError("n must be a non-negative integer.")

    total = 0
    fact = 1  # factorial acumulado: en cada paso paso de (i-1)! a i! haciendo fact *= i

    for i in range(1, n + 1):
        fact *= i      # ahora fact = i!
        total += fact  # acumulo i!

    return total


# factorial_sum(n) (versión original) ya está definida en 3.1
for n_test in [0, 1, 5, 10, 50, 100]:
    assert factorial_sum(n_test) == factorial_sum_optimized(n_test)

print("Comprobación pasada: ambas versiones devuelven el mismo resultado para los casos de prueba.")


n_test = 3500  # mismo valor que en el ejercicio 3.1 para comparar

print("\nTiempo con la versión original (O(n^2)):")
%time _ = factorial_sum(n_test)  # asigno el resultado a _ para no imprimir el entero gigante

print("\nTiempo con la versión optimizada (O(n)):")
%time _ = factorial_sum_optimized(n_test)
`````

`````python
import multiprocessing as mp

def factoriales_secuencial(valores: list[int]) -> list[int]:
    """
    Calculates factorials of a list of integers in a sequential way.

    Args:
        valores (list[int]): list of integers.

    Returns:
        list[int]: list with the factorial of each integer in the input.
    """
    inicio = time.perf_counter()

    resultados: list[int] = []
    for n in valores:
        # calculo uno a uno, sin ninguna magia, para que quede claro lo que pasa
        resultados.append(calcular_factorial(n))

    fin = time.perf_counter()
    duracion = fin - inicio
    print(f"Tiempo total (secuencial): {duracion:.2f} s")

    return resultados


def factoriales_multiproceso(valores: list[int]) -> list[int]:
    """
    Calculates factorials of a list of integers using multiple processes.

    Args:
        valores (list[int]): list of integers.

    Returns:
        list[int]: list with the factorial of each integer in the input.
    """
    inicio = time.perf_counter()

    # uso un Pool para repartir el trabajo entre varios procesos
    with mp.Pool() as pool:
        resultados: list[int] = pool.map(calcular_factorial, valores)

    fin = time.perf_counter()
    duracion = fin - inicio
    print(f"Tiempo total (multiproceso): {duracion:.2f} s")

    return resultados
`````

### ¿Es igual de eficiente *multiprocessing* para valores pequeños y grandes de *n*?

En este problema el cálculo del factorial es una tarea claramente *CPU-bound*: casi todo el tiempo se dedica a cómputo, no a esperas de entrada/salida. Justo en este tipo de tareas es donde la Unidad 5 indica que el multiproceso puede aportar mejoras importantes, al repartir el trabajo entre varias CPUs físicas.

Con los valores grandes de `n` que uso aquí (150 000, 200 000, …, 400 000), sí se aprecia una mejora: el tiempo de pared de la versión multiproceso es sensiblemente menor que el de la versión secuencial. Sin embargo, si utilizase valores de `n` mucho más pequeños (o una lista con muy pocos elementos), es probable que el multiproceso no fuera más rápido e incluso saliera peor. Cuando el trabajo que hace cada proceso es pequeño, el tiempo que se ahorra en cómputo no compensa el coste de crear el *pool*, repartir los datos y recoger los resultados.

En resumen:

* Para factoriales grandes y listas con un tamaño razonable, `multiprocessing` puede reducir el tiempo total.
* Para factoriales pequeños o listas muy cortas, el beneficio se reduce mucho o desaparece, porque el *overhead* de gestionar procesos adicionales domina el coste total.

### Factores que pueden hacer que la ejecución paralela sea más lenta que la secuencial

Aunque la intuición pueda ser “más procesos ⇒ más rápido”, en la práctica hay varios factores que pueden hacer que una versión paralela resulte incluso más lenta que la secuencial:

* **Overhead de creación de procesos.**
  Crear un proceso nuevo implica reservar memoria, copiar parte del estado y arrancar el intérprete de Python. Este coste es bastante mayor que el de crear un *thread*. Si el trabajo de cada proceso es pequeño, este overhead puede dominar.

* **Coste de comunicación y serialización de datos.**
  En Python, los procesos no comparten memoria; para pasar datos entre ellos hay que serializarlos (*pickling*) y enviarlos por *pipes* o colas. Cuando los argumentos o resultados son grandes, el tiempo de copiado puede comerse buena parte de la ganancia.

* **Número limitado de CPUs.**
  Si se lanzan más procesos que núcleos disponibles, el sistema operativo tiene que hacer más cambios de contexto para repartir la CPU entre ellos, lo que añade overhead y puede reducir la eficiencia global.

* **Contención de recursos.**
  En otros programas pueden aparecer cuellos de botella en memoria, disco, red, etc. Muchos procesos accediendo al mismo recurso pueden introducir más esperas que una versión secuencial bien organizada.

En la Unidad 5 se insiste en esta idea: no todos los programas se benefician de la concurrencia o el paralelismo; hay casos en los que la versión con hilos o procesos es igual o incluso más lenta que la secuencial, sobre todo cuando la carga de trabajo por tarea es pequeña o cuando el overhead de gestionar los hilos/procesos es demasiado grande para el problema que se está resolviendo.

Esto encaja con la explicación teórica: para aprovechar realmente `multiprocessing` hacen falta cálculos pesados y un volumen de datos suficiente, de modo que el tiempo de cómputo compense el coste añadido de crear y coordinar los procesos.

Una empresa dedicada a la creación de contenidos digitales os ha solicitado desarrollar una galería de imágenes obtenidas desde una API pública. El objetivo del proyecto es descargar las imágenes y almacenarlas localmente para su posterior uso dentro de una aplicación interna de gestión y visualización.

En una primera fase del desarrollo, se han recopilado diversas URLs de imágenes y se busca evaluar la eficiencia del proceso de descarga bajo distintos enfoques de ejecución.

Para ello, podéis acceder al listado de imágenes a descargar llamando a la función `get_image_urls()`, que devuelve una lista de URLs de una API pública. Además, se proporciona un ejemplo de función para descargar una imagen individual, suponiendo que ya existe la carpeta `imgs/` donde se almacenarán los archivos descargados.

`````python
import requests

def get_image_urls(num: int = 50) -> list[str]:
    """
    Fetches a list of random image URLs from the API.

    Args:
        num (int): Number of image URLs to retrieve.

    Returns:
        list[str]: A list containing the image URLs.
    """
    api_url = f"https://picsum.photos/v2/list?page=1&limit={num}"
    response = requests.get(api_url)
    return [item["download_url"] for item in response.json()]
`````

`````python
import os

def download_image(url: str, idx: int) -> None:
    """
    Downloads a single image from a given URL and
    saves it with a unique index.

    Args:
        url (str): The URL of the image to download.
        idx (int): The index used to create a unique filename.
    """
    filename = os.path.join("imgs", f"image_{idx}.jpg")
    try:
        r = requests.get(url)
        with open(filename, "wb") as f:
            f.write(r.content)
        print(f"Saved: {filename}")
    except requests.RequestException as e:
        print(f"Error downloading {url}: {e}")
`````

### Elección para este problema: solución concurrente con hilos

En el caso de la descarga de imágenes desde una API, la mayor parte del tiempo no se dedica a realizar cálculos, sino a **esperar**: esperar a que la petición HTTP salga por la red, esperar a que el servidor responda y a que lleguen los datos. Es, por tanto, una tarea **I/O-bound** (limitada por la entrada/salida de red) y no **CPU-bound**.

En la Unidad 5 se explica que, en este tipo de tareas donde el cuello de botella es la red y el uso de CPU es mínimo, una solución con **múltiples hilos** permite solapar esas esperas: mientras un hilo está bloqueado esperando la respuesta del servidor, otro hilo puede lanzar otra petición o procesar otra respuesta. Además, crear hilos suele tener **menos overhead** que crear procesos nuevos, porque los hilos comparten el mismo espacio de memoria del proceso principal.   

Por ese motivo, para este problema concreto considero más adecuado implementar una solución **concurrente multithreaded** (por ejemplo, varios hilos que llaman a `download_image` en paralelo), antes que una solución basada en múltiples procesos. El uso de varios procesos no aporta una mejora clara, porque el trabajo de CPU por imagen es muy pequeño y el tiempo total sigue estando dominado por la red, tal y como también se comenta en el ejemplo final de la Unidad 5 con llamadas a una API remota.

En resumen, para **descargar muchas imágenes desde Internet**, mi elección sería:

- Priorizar **multithreading** para solapar esperas de red con un overhead moderado.
- Reservar `multiprocessing` solo si, además de descargar, tuviera que hacer cálculos intensivos sobre cada imagen (por ejemplo, aplicar filtros pesados, reconocimiento de patrones, etc.).

### ¿Cuándo usaría multiprocessing (estrategia no elegida aquí)?

Aunque en este problema concreto prefiero una solución concurrente con hilos, la estrategia de **multiprocessing** es más apropiada en otros contextos, especialmente cuando:

- La tarea es claramente **CPU-bound** y se puede descomponer en subtareas independientes.
- Quiero aprovechar **varios núcleos físicos** de la máquina para acelerar el cálculo.
- El coste de crear procesos y comunicar datos entre ellos se compensa con el ahorro de tiempo en los cálculos.

Algunos ejemplos de situaciones donde usaría `multiprocessing` serían:

- Calcular descomposiciones en factores primos, grandes factoriales u otras operaciones matemáticas pesadas para muchos valores, como se ilustra en los ejemplos de la Unidad 5.
- Aplicar algoritmos de **procesamiento de imágenes** muy costosos (detección de bordes, convoluciones grandes, compresión, etc.) sobre miles de archivos.
- Entrenar varios modelos de *machine learning* en paralelo (por ejemplo, validación cruzada con distintos hiperparámetros), repartiendo modelos en diferentes procesos.

En estos escenarios, la ejecución paralela en varios procesos sí aprovecha de verdad los núcleos de la CPU y puede reducir de forma importante el tiempo total de cálculo, mientras que una versión multithreaded quedaría limitada por el *Global Interpreter Lock* y no sería tan efectiva.

   tiempo total de descarga en el eje Y), junto con una línea horizontal que marca el tiempo

Después interpreto cómo cambia el rendimiento al aumentar el número de hilos o procesos y comento

(como las descargas por red) lo esperable es que el multithreading tenga bastante sentido, mientras
que el multiprocessing puede añadir más overhead de creación de procesos y comunicación.


`````python
import os
import time
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor


def download_images_sequential_from_urls(urls: list[str]) -> float:
    """
    Downloads all images in the given list of URLs sequentially
    and returns the total execution time in seconds.
    """
    os.makedirs("imgs", exist_ok=True)

    inicio = time.perf_counter()

    for idx, url in enumerate(urls, start=1):
        # reutilizo la funcion proporcionada en el enunciado
        download_image(url, idx)

    fin = time.perf_counter()
    return fin - inicio


def download_images_threadpool(urls: list[str], n_workers: int) -> float:
    """
    Downloads all images using a ThreadPoolExecutor with n_workers threads.
    Returns the total execution time in seconds.
    """
    os.makedirs("imgs", exist_ok=True)

    inicio = time.perf_counter()

    # lanzo varios hilos, cada uno ejecutando download_image sobre una url
    with ThreadPoolExecutor(max_workers=n_workers) as executor:
        futures = [
            executor.submit(download_image, url, idx)
            for idx, url in enumerate(urls, start=1)
        ]
        # espero a que terminen todos los hilos
        for fut in futures:
            fut.result()

    fin = time.perf_counter()
    return fin - inicio


def download_images_processpool(urls: list[str], n_workers: int) -> float:
    """
    Downloads all images using a ProcessPoolExecutor with n_workers processes.
    Returns the total execution time in seconds.
    """
    os.makedirs("imgs", exist_ok=True)

    inicio = time.perf_counter()

    # igual que con los hilos, pero repartiendo el trabajo entre varios procesos
    with ProcessPoolExecutor(max_workers=n_workers) as executor:
        futures = [
            executor.submit(download_image, url, idx)
            for idx, url in enumerate(urls, start=1)
        ]
        for fut in futures:
            fut.result()

    fin = time.perf_counter()
    return fin - inicio


def medir_tiempos_concurrentes(num_images: int = 50, max_workers: int = 10):
    """
    Measures sequential, threaded and multiprocess download times
    for a given number of images and workers.

    Returns:
        workers (list[int]): list of numbers of workers tested.
        tiempos_threads (list[float]): times with ThreadPoolExecutor.
        tiempos_procs (list[float]): times with ProcessPoolExecutor.
        tiempo_secuencial (float): baseline sequential time.
    """
    # obtengo una sola vez la lista de URLs para que todos los casos usen el mismo conjunto
    urls: list[str] = get_image_urls(num_images)

    print(f"Descargando {num_images} imagenes para cada configuracion...\n")

    # tiempo base secuencial
    tiempo_secuencial = download_images_sequential_from_urls(urls)
    print(f"Tiempo secuencial base: {tiempo_secuencial:.2f} s\n")

    workers = list(range(1, max_workers + 1))
    tiempos_threads: list[float] = []
    tiempos_procs: list[float] = []

    for n in workers:
        print(f"== {n} workers ==")
        t_threads = download_images_threadpool(urls, n)
        t_procs = download_images_processpool(urls, n)
        tiempos_threads.append(t_threads)
        tiempos_procs.append(t_procs)
        print(
            f"Threads ({n}): {t_threads:.2f} s | "
            f"Processes ({n}): {t_procs:.2f} s\n"
        )

    return workers, tiempos_threads, tiempos_procs, tiempo_secuencial
`````

`````python
import matplotlib.pyplot as plt

# ejecuto la comparativa y dibujo la grafica
if __name__ == "__main__":
    workers, tiempos_threads, tiempos_procs, tiempo_secuencial = medir_tiempos_concurrentes()

    # resumen por consola
    for n, t_th, t_pr in zip(workers, tiempos_threads, tiempos_procs):
        print(
            f"workers = {n:2d} | "
            f"ThreadPoolExecutor = {t_th:.2f} s | "
            f"ProcessPoolExecutor = {t_pr:.2f} s"
        )

    # grafica de comparacion
    plt.figure(figsize=(8, 5))
    plt.plot(workers, tiempos_threads, marker="o", label="ThreadPoolExecutor (hilos)")
    plt.plot(workers, tiempos_procs, marker="o", label="ProcessPoolExecutor (procesos)")
    plt.axhline(
        y=tiempo_secuencial,
        linestyle="--",
        label=f"Secuencial (≈ {tiempo_secuencial:.2f} s)",
    )
    plt.xlabel("Número de hilos / procesos")
    plt.ylabel("Tiempo total de descarga (s)")
    plt.title("Descarga de 50 imágenes: comparación de tiempos")
    plt.grid(True)
    plt.legend()
    plt.show()
`````
